```python
def decode(data):
    parts = data.split("_")
    if len(parts) == 3:
        return pd.Series(parts) 
    else:
        p1 = parts.pop()
        parts.extend(["not_specified", p1])
        return pd.Series(parts)

df_long[['casetype', 'diseasetype', 'sexage']] = df_long['code'].apply(decode)
```

```python
def decode2(data):
    gender= 'male' if data[0] == 'm' else 'female'
    a=data[1:]
    if len(a) ==4:
        p1=a[:2]
        p2=a[2:]
    elif len(a) ==3:
        p1=a[0]
        p2=a[1:]
    else:
        p1="greater then "
        p2=a[:]
    age_group=f"{p1}-{p2}"
    parts=[gender,age_group]
    return pd.Series(parts) 
df_long[['gender', 'age_group']]=df_long['sexage'].apply(decode2)
```
