```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv(r"D:\Data_Science\datasets\netflix\netflix_titles.csv")
```

```python
#Medium: Plot the distribution of content types (Movies vs TV Shows).
df1=df.groupby('type')['show_id'].count()
plt.pie(df1.values,labels=df1.index,startangle=90,autopct='%1.1f%%',colors=['green','yellow'],shadow=True,wedgeprops={'edgecolor':'black'},explode=[0,0.1])
plt.title("distribution of content types (Movies vs TV Shows).")
plt.show()
```

```python
#Difficult: For a given year, plot a pie chart showing the distribution of ratings.
df=df.dropna(subset='date_added')
df=df.dropna(subset='rating')
df['Year']=df['date_added'].str.split(",").str[1].str.strip().astype('Int16')
df=df[~df.rating.str.contains(r"((\d)+ min)")]
df1=df.groupby(['Year','type']).size()
info1=df1.xs("Movie",level='type')
info2=df1.xs("TV Show",level='type')
index_all=info1.index.union(info2.index)
info11=info1.reindex(index_all,fill_value=0)
info12=info2.reindex(index_all,fill_value=0)
d1=dict()
x=2015
d1["Movie"]=info11[x]
d1["TV Show"]=info12[x]
plt.pie(d1.values(),labels=d1.keys(),startangle=90,autopct='%1.1f%%',colors=['green','yellow'],shadow=True,wedgeprops={'edgecolor':'black'},explode=[0,0.1])
plt.title(f"distribution of content types  (Movies vs TV Shows) Year : {x}")
plt.show()
```
