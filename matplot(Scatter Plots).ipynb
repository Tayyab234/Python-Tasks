```python
df=df.dropna(subset='rating')
df=df.dropna(subset=['duration'])
df_fil=df[~(df['duration'].str.contains(r"Season"))]
# unique ratings
unique_ratings = df['rating'].dropna().unique()
# generate a color map
colors = plt.cm.tab10(range(len(unique_ratings)))  # tab10 is a good categorical palette
# build a dict
rating_color = dict(zip(unique_ratings, colors))
df_fil['duration']=df_fil['duration'].str.split().str[0].astype("Int16")
plt.scatter(df_fil['release_year'],df_fil['duration'],edgecolor='black',alpha=0.25,c=df_fil['rating'].map(rating_color))
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# drop rows with missing rating or duration
df_clean = df.dropna(subset=['rating', 'duration'])

# keep only movies (duration in minutes)
df_movies = df_clean[~df_clean['duration'].str.contains("Season")].copy()

# convert duration to minutes (int)
df_movies['duration_min'] = df_movies['duration'].str.extract(r'(\d+)').astype(float)

# get unique ratings
unique_ratings = df_movies['rating'].dropna().unique()
colors = plt.cm.get_cmap('tab10', len(unique_ratings))

# build a color dictionary
rating_color = dict(zip(unique_ratings, colors))

# map colors
color_mapped = df_movies['rating'].map(rating_color)

# jitter for x and y to reduce overlapping
jitter_x = df_movies['release_year'] + np.random.uniform(-0.3, 0.3, size=len(df_movies))
jitter_y = df_movies['duration_min'] + np.random.uniform(-2, 2, size=len(df_movies))

# plot
plt.figure(figsize=(12, 6))
scatter = plt.scatter(
    jitter_x,
    jitter_y,
    c=color_mapped,
    edgecolor='black',
    alpha=0.5
)

# make a legend
from matplotlib.lines import Line2D
legend_elements = [
    Line2D([0], [0], marker='o', color='w', markerfacecolor=rating_color[r], 
           label=r, markersize=8, markeredgecolor='black')
    for r in unique_ratings
]
plt.legend(handles=legend_elements, bbox_to_anchor=(1.02, 1), loc='upper left', title="Rating")

plt.title("Movie Release Year vs Duration (with Jitter, Colored by Rating)")
plt.xlabel("Release Year")
plt.ylabel("Duration (minutes)")
plt.grid(True)
plt.tight_layout()
plt.show()
```
