```python
# Custom Exception
class InvalidElementError(Exception):
    """Raised when an invalid element is added to the set-like structure."""
    pass
class error1(Exception):
    pass
class EvenIntSet:
    def __init__(self):
        self.data = set()

    def validate(self, *elements):
        """Validate that all elements are even integers."""
        for elem in elements:
            if not isinstance(elem, int) or elem % 2 != 0:
                raise InvalidElementError(f"Invalid element: {elem} (Only even integers allowed)")

    def add(self, *elements):
        """Add one or multiple elements."""
        self.validate(*elements)
        self.data.update(elements)

    def remove(self, *elements):
        """Remove one or multiple elements."""
        for elem in elements:
            if elem not in self.data:
                raise KeyError(f"Element {elem} not found in the set.")
            self.data.remove(elem)

    def union(self, other):
        if not isinstance(other, EvenIntSet):
            raise TypeError("Can only perform union with another EvenIntSet")
        new_set = EvenIntSet()
        new_set.data = self.data.union(other.data)
        return new_set

    def intersection(self, other):
        if not isinstance(other, EvenIntSet):
            raise TypeError("Can only perform intersection with another EvenIntSet")
        new_set = EvenIntSet()
        new_set.data = self.data.intersection(other.data)
        return new_set

    def difference(self, other):
        if not isinstance(other, EvenIntSet):
            raise TypeError("Can only perform difference with another EvenIntSet")
        new_set = EvenIntSet()
        new_set.data = self.data.difference(other.data)
        return new_set

    def __str__(self): #use of print operator in object
        return f"EvenIntSet({self.data})"

    def __len__(self):#use of len operator in object
        return len(self.data)

    def __contains__(self, item):  #use of in operator in object
        return item in self.data
        
    def __eq__(self, other):
        if not isinstance(other, EvenIntSet):
            raise error1("Comparison must be with an EvenIntSet")
        return self.data == other.data

    def __ne__(self, other):
        if not isinstance(other, EvenIntSet):
            raise error1("Comparison must be with an EvenIntSet")
        return self.data != other.data
    
    def __le__(self, other):
        if not isinstance(other, EvenIntSet):
            raise error1("Comparison must be with an EvenIntSet")
        return self.data <= other.data
    
    def __add__(self, other):
        if not isinstance(other, EvenIntSet):
            raise error1("Comparison must be with an EvenIntSet")
        return  self.data.union(other.data)
        
    def __iter__(self):  #use of loop in object
        return iter(self.data)
a = EvenIntSet()
a.add(2, 4, 6)

b = EvenIntSet()
b.add(4, 6, 8)

c = a.union(b)
print(c)  # EvenIntSet({2, 4, 6, 8})

d = a.intersection(b)
print(d)  # EvenIntSet({4, 6})

e = a.difference(b)
print(e)  # EvenIntSet({2})

a.remove(2)
print(a)  # EvenIntSet({4, 6})
print(6 in a,len(a))
print(a+b)
# Test Exception
try:
    a.add(3, 4)  # Should raise InvalidElementError
except InvalidElementError as e:
    print(e)
```

```python
from collections import namedtuple, defaultdict

# Sample Sales Record
SalesRecord = namedtuple('SalesRecord', ['customer', 'date', 'amount'])

# Example sales data
records = [
    SalesRecord('Alice', '2024-06-01', 250),
    SalesRecord('Bob', '2024-06-03', 400),
    SalesRecord('Alice', '2024-06-02', 300),
    SalesRecord('Bob', '2024-06-01', 200),
    SalesRecord('Charlie', '2024-06-02', 500)
]

# Sort by date first, then by amount
sorted_records = sorted(records, key=lambda x: (x.date, x.amount))

# Group by customer and calculate total sales
#customer_totals = dict()
customer_totals = defaultdict(int)#default value zero
for record in sorted_records:
    customer_totals[record.customer] += record.amount

print("Sorted Records:")
for r in sorted_records:
    print(r)

print("\nTotal Sales per Customer:")
for customer, total in customer_totals.items():
    print(f"{customer}: {total}")
```

```python
def extract_paths(d, current_path=()):
    paths = []
    for key, value in d.items():
        new_path = current_path + (key,)
        if isinstance(value, dict):
            paths.extend(extract_paths(value, new_path))#like a+ b but it extend
        else:
            paths.append(new_path)
    return paths

# Example nested dictionary
nested_dict = {
    "a": {"b": 1},
    "c": {"d": {"e": 2}}
}

key_paths = extract_paths(nested_dict)
print("Key Paths:")
print(key_paths)
```

```python
# Example dictionary
data = {'a': {1, 2}, 'b': {2, 3}, 'c': {3, 4}}

# Invert dictionary: Each number should point to all keys that contain it
inverted = {}

for key, values in data.items():
    for value in values:
        if value not in inverted:
            inverted[value] = set()
        inverted[value].add(key)

print("Inverted Dictionary:")
for val, keys in inverted.items():
    print(f"{val}: {keys}")
```
