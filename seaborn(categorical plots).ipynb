```python
df1=sns.load_dataset("titanic")
df1.head(2)
```

```python
df1.info()
```

```python
# 8Ô∏è‚É£ pointplot (titanic)
# üëâ ‚ÄúPlot survival rate by class with pointplot‚Äù
# Step 1: Group and calculate survival rate
h = df1.groupby("class")['survived'].value_counts()
k1 = h.xs(1, level='survived')  # survivors
k0 = h.xs(0, level='survived')  # non-survivors

# Step 2: Compute survival rate
survival_rate = (k1 / (k1 + k0)).reset_index()
survival_rate.columns = ['class', 'survival_rate']

# Step 3: Point plot
#sns.pointplot(data=survival_rate, x='class', y='survival_rate')
plt.plot(survival_rate['class'],survival_rate['survival_rate'],marker='o')
plt.title("Survival Rate by Class")
plt.ylabel("Survival Rate")
plt.ylim(0, 1)
plt.show()
```

```python
sns.pointplot(data=df1,x='class',y='survived',ci=None)
```
