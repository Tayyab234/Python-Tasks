```python
df=sns.load_dataset("penguins")
df.head(2)
```

```python
df.info()
```

```python
#  scatterplot (penguins)
# 👉 “Plot bill length vs bill depth, colored by species”
un=df['species'].unique()
df=df.dropna(subset=['bill_length_mm','bill_depth_mm'])
sns.scatterplot(data=df,x='bill_length_mm',y='bill_depth_mm',hue='sex',style='species')
```

```python
#  scatterplot (penguins)
# 👉 “Plot bill length vs bill depth, colored by species”
un=df['species'].unique()
df=df.dropna(subset=['bill_length_mm','bill_depth_mm'])
sns.relplot(data=df,x='bill_length_mm',y='bill_depth_mm',kind='scatter',hue='species',style='sex',size='body_mass_g')#sizes=(90,100) min max size controller
```

```python
sns.relplot(
    data=df,
    x='bill_length_mm',
    y='bill_depth_mm',
    hue='species',
    style='sex',
    kind='scatter',
    s= 100  
)
#plt.savefig(r"D:/Data_Science")
```

```python
df2=sns.load_dataset("flights")
df2.head(2)
```

```python
df2.info()
```

```python
# 2️⃣ lineplot (flights)
# 👉 “Plot number of passengers over time, per year”
# s1=df2.groupby('year')['passengers'].sum().sort_index()
# sns.lineplot(data=df2,x=s1.index,y=s1.values)
s1 = df2.groupby('year')['passengers'].sum().sort_index()
df_grouped = s1.reset_index(name='total_passengers')

sns.lineplot(data=df_grouped, x='year', y='total_passengers',marker='o')  # you can pass dataframe only
# plt.title('Total Passengers per Year')
# plt.ylabel('Passengers')
# plt.show()
```
